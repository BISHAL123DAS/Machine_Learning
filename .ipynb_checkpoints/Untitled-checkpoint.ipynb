# Detecting outlier using z score
using z score 
formula for z score=(observation-mean standard deviation)
z=(X-u)/Q 

```python
outliers=[]
def detect_outliers(data):
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers        
```

```python
outlier_pts = detect_outliers(dsa1)
```
